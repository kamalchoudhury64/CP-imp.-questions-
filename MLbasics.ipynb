```
import pandas as pd
import io
movie_train_data = pd.read_csv('movies.csv')
movie_train_data
```

```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
# Removing extra information
## dropna is a function used to drop the rows/columns with the Nan/Null values. 
## Function is described as follows --> Dataframe.dropna(axis = ( 0/1 --> column/row ) , how = ('any'/'all') --> (all/any) values are NULL that row/column is removed , thresh = num --> minimum number of na values to drop)    
movie_train_data = movie_train_data.dropna(axis = 0,how = 'any')
X = movie_train_data.drop(columns = ['Rank','Title','Genre','Director','Revenue','Description']) # Here X means the input file
y = movie_train_data['Revenue'] # y is representing the Revenue earned 

# Training the complete Data and predicting our inputs
model = DecisionTreeClassifier()
model.fit(X,y)
# Getting predictions and storing in a list 
predictions = model.predict([['2000','8.9','90','1201201','200'],['2004','6.9','90','2601201','150']])
print(predictions)
```

```
# Splitting the complete data into training and testing sets and calculating accuracy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.56 )
print(X_train)
print(X_test)
# The train_test_split function returns a tuple with 1 and 2 as training and test sets for input file and 3 and 4 as train and test for output
model1 = DecisionTreeClassifier()
# Training the models based on splitted arrays
model1.fit(X_train,y_train)
predictions = model1.predict(X_test)
print("Predictions are \n",predictions)
print("Actual Values are \n",y_test)
win = accuracy_score(predictions,y_test)
print("Accuracy is : ",win)
```
